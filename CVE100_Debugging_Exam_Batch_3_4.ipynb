```python
class Student:
    def __init__(self, name, scores, score):
        self.name = name
        self.scores = scores
        self.score = score

    def add_score(self, score):
        self.scores.add(scores)

    def avg(self):
        avg_score = sum(self.scores) / len(self.scores)
        return avg_score
    def grade(self):
        avg_score = sum(self.scores) / len(self.scores)
          
        if avg_score >= 80:
          grade = 'A'
        elif avg_score >= 80:
          grade = 'B'
        else :
          grade = 'C'
        return grade


students = [
    Student("A", [80, 70, 90], 85),
    Student("B", [60, 50, 65], 55)
]

for s in students:
    print(s.name, "average:", s.avg())
    print(s.name, "grade:", s.grade())

# Expected Output:
# A average: 80.0
# A grade: A
# B average: 58.33
# B grade: B
```

```python
def write_data(filename):
    f = open(filename, "r")
    f.write("10 20 30 40")
    f.close()

def read_and_process(filename):
    with open(filename, "w") as f:
        data = f.read().split()
    total = 0
    for x in data:
        total += x
    avg = sum(total) / len(total)
    return avg

write_data("nums.txt") #file should be appeared on your computer
print("Average:", read_and_process("num.txt"))
# Expected Output: Average: 25.0
```
