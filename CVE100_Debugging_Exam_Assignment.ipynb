```python
def average_by_key(rows):
    if len(rows) == 0:   
        return 0.0

    total = 0
    for r in rows:
        total += r["score"]   

    return total / len(rows)
print(average_by_key([{"score": 5}, {"score": 7}]) ) # Expected Output: 6.0/ 2.5
print(average_by_key([]))                            # Expected Output: 0.0(NONE)
```

```python
def even_numbers(nums):
    evens = []             
    for n in nums:         
        if n % 2 == 0:
            evens.append(n)
    return evens           

print(even_numbers([1,2,3,4,5,6])) 
```

```python
class Artist:
    def __init__(self, name, fee):
        self.name = name
        self.__fee = fee  #private attribute, do not change this.

    def add_charges(self, additional_charges):
        self.__fee += additional_charges

    @property
    def fee(self):        
        return self.__fee


a1 = Artist("Taylor Swift", 50)
a2 = Artist("Lisa", 65)

a1.add_charges(20)
a2.add_charges(20)

print("Taylor Swift fee=", a1.fee)  # Expected Output: 70
print("Lisa fee=", a2.fee)  # Expected Output: 85
```
