```python
# Expected Output:
# Sum of first 3 positives: 12

nums = [-1, 2, -5, 3, 7, -2, 4]
count = 0
total = 0

for n in nums:
    if n <= 0:
        continue
    total += n
    count += 1
    if count == 3:
        break
    else:
        continue

print("Sum of first 3 positives:", total)
```

```python
# find the highest scoring player
# Expected Output:
# Winner: Cara
players = [("Alice", 15), ("Bob", 10), ("Cara", 20), ("Dan", 9)]
winner = None
best = -1

for name, score in players:
    if score > best:
        best = score
        print("best score=", best)
        winner = name

print("Winner:", winner)
```

```python
# Expected Output:
# First positive multiple of 7: 7
nums = [-14, -7, 0, 7, 14, 21, 28]
first = None
for n in nums:
    if n <= 0:
        continue
    if n % 7 == 0:
        first = n
        break
    else:
        continue

if first is None:
    print("None found")
else:
    print("First positive multiple of 7:", first)
```

```python
# Expected Output:
# mean=3.0 min=1 max=5

def stats(nums):
    mn = min(nums)
    mx = max(nums)
    avg = sum(nums)/len(nums)
    return mn, mx, avg

mn, mx, avg = stats([1,2,3,4,5])
print(f"mean={avg} min={mn} max={mx}")
```

```python
# Expected Output:
# Brand: Toyota, Engine: 2000
class Vehicle:
    def __init__(self, brand):
        self.brand = brand

class Car(Vehicle):
    def __init__(self, brand, engine):
        super().__init__(brand)
        self.engine = engine

c = Car("Toyota", 2000)
print(f"Brand: {c.brand}, Engine: {c.engine}")
```

```python
# Expected Output:
# Area: 12.56
import math

class Shape:
    def area(self):
        pass

class Circle(Shape):
    def area(self, radius):
        return math.pi * radius ** 2

c = Circle()
print("Area:", c.area(2))
```
