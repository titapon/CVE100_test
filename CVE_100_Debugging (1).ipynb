**Exercise 1: Find the Misspelled Variable**

```python
def calculate_total_price(price, quantity):
    """Calculate total price including tax"""
    total = price * quantity
    tax_rate = 0.08
    total_with_tax = total + (total * tax_rate)
    return total_with_tax
```

```python
result = calculate_total_price(10, 5)
print(f"Total price: {result:.2f}")
```

**Exercise 3: Index Error**

```python
def get_first_and_last(items):
    """Return the first and last items from a list"""
    first = items[0]
    last = items[3]
    return first, last

colors = ["red", "blue", "green", "yellow"]
first_color, last_color = get_first_and_last(colors)
print(f"First: {first_color}, Last: {last_color}")
```

**Exercise 8: Division by Zero**

```python
def calculate_batting_average(hits, at_bats):
    """Calculate batting average in baseball"""
    average = hits / at_bats
    return round(average, 3)

avg = calculate_batting_average(1, 3)
print(f"Batting average: {avg}")
```

**Exercise 11: Infinite Loop**

```python
def countdown(n):
    """Count down from n to 1"""
    while n > 0:
        print(n)
        n -= 1


countdown(5)
```

```python
# Calculate the total cost of items with tax
def calculate_total(prices, tax_rate):
    subtotal = 0

    for price in prices:
        subtotal = subtotal + price

    tax = subtotal * tax_rate / 100
    total = subtotal - tax

    return total

# Test the function
items = [10.00, 25.50, 5.75]
total = calculate_total(items, 8)
print(f"Total cost: ${total}")
```

```python
def  calculate_totals(sales_list, expenses_list):
    total = 0
    expense_total = 0
    for sale in sales_list:
        total += sale
    print(f"Sales total: {total}")
    
    for expense in expenses_list:
        expense_total += expense
    print(f"Expenses total: {expense_total}")
        
```

```python
sales = [400, 300, 1000, 500]
expenses = [20, 30, 50, 20]
print(calculate_totals(sales, expenses))

#expected output
#Sales Total = 2200
#Expenses Total = 120
```
